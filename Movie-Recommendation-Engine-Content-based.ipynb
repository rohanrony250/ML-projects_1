```python
import pandas as pd
from math import sqrt
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
movies_df = pd.read_csv('./Movie-Recommendation-Dataset/movies_metadata.csv')
ratings_df = pd.read_csv('./Movie-Recommendation-Dataset/ratings.csv')
movies_df.head()
```

```python
#finding columns that have missing values :

missing_data = movies_df.isnull()
missing_data.head()    
```

```python
# counting no:of NaN values in each column so as to have a better insight on what to drop


for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts(), "\n")
    
#from here we find that columns 'homepage' & 'belongs_to_collection' has most of its values as NaN so it can be avoided for our prediction.
#other columns with missing data can either have its respective empty rows dropped or replaced with frequency/mean/other-functions, whichever method seems fit.
```

```python
#drop homepage and collection column 

movies_df = movies_df.drop(["homepage", "belongs_to_collection", "tagline"], axis=1)
```

```python
movies_df.dropna(subset = ["release_date"], axis = 0, inplace = True)
movies_df.reset_index(drop=True, inplace = True)
```
